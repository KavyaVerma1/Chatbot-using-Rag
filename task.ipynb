```python
import pinecone
from langchain_huggingface import HuggingFaceEmbeddings
from langchain.vectorstores import Pinecone
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
# Initialize the Hugging Face embeddings model
embeddings_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")

# Step 1: Read the Text File
def read_text_file(file_path):
    with open(file_path, "r", encoding="utf-8") as file:
        text = file.read()
    return text

# Step 2: Split the Text into Chunks
def split_text_into_chunks(text, chunk_size=500, chunk_overlap=50):
    splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
    chunks = splitter.split_text(text)
    return chunks

# Step 3: Generate Embeddings for Chunks
def generate_embeddings(chunks):
    return [embeddings_model.embed_query(chunk) for chunk in chunks]

# Example Usage:
file_path = "/content/Harry_Potter_all_books_preprocessed.txt"

# Read and process text file
text = read_text_file(file_path)
chunks = split_text_into_chunks(text)
embeddings = generate_embeddings(chunks)

# Print the result
print("Number of chunks created:", len(chunks))
print("First chunk embedding:", embeddings[0])
```

```python
from pinecone import Pinecone
pinecone_api_key = userdata.get('PINECONE_API_KEY')
# configure client
pc = Pinecone(api_key=pinecone_api_key)
```

```python
import os
from pinecone import ServerlessSpec

cloud = os.environ.get('PINECONE_CLOUD') or 'aws'
region = os.environ.get('PINECONE_REGION') or 'us-east-1'

spec = ServerlessSpec(cloud=cloud, region=region)
```

```python
index_name = 'harry-potter'
```

```python
import time

if index_name in pc.list_indexes().names():
    pc.delete_index(index_name)

# we create a new index
pc.create_index(
        index_name,
        dimension=768,  # dimensionality of text-embedding-ada-002
        metric='dotproduct',
        spec=spec
    )

# wait for index to be initialized
while not pc.describe_index(index_name).status['ready']:
    time.sleep(1)
```

```python
index = pc.Index(index_name)
# wait a moment for connection
time.sleep(1)

index.describe_index_stats()
```

```python
# Upsert the embeddings into Pinecone in batches
BATCH_SIZE = 100  # Define the batch size for uploading

# Prepare the embeddings and metadata for upsert
def create_batches(embeddings, chunks, batch_size=BATCH_SIZE):
    for i in range(0, len(embeddings), batch_size):
        # Create a batch of vectors (id, embedding, metadata)
        batch = [
            (f"chunk_{i+j}", embeddings[i+j], {"text": chunks[i+j], "chunk_number": i+j})
            for j in range(min(batch_size, len(embeddings) - i))
        ]
        yield batch

# Uploading the embeddings in batches to Pinecone
for batch in create_batches(embeddings, chunks):
    index.upsert(vectors=batch)

print("Embeddings have been successfully upserted to Pinecone.")
```

```python
import os
import pinecone
from langchain.vectorstores import Pinecone
from langchain.embeddings import HuggingFaceEmbeddings
from pinecone import ServerlessSpec

# Initialize the Hugging Face embeddings model (MPNet)
embed = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")

text_field = "text"

# Set cloud and region from environment or defaults
cloud = os.environ.get('PINECONE_CLOUD') or 'aws'
region = os.environ.get('PINECONE_REGION') or 'us-east-1'

# Create a ServerlessSpec object with the cloud and region
spec = ServerlessSpec(cloud=cloud, region=region)



# Create the vector store with the index
vectorstore = Pinecone(
    index, embed.embed_query, text_field
)

print("Vectorstore successfully created with MPNet embeddings.")
```
